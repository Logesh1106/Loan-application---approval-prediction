```python
# 7.Create a simple predictive model to assess whether a loan application will be approved or rejected and provide the accuracy score.
column=['Gender','Married','Dependents','Education','Self_Employed','Credit_History','Property_Area','Income','Application_Status']
all= LabelEncoder()
for i in column:
    df[i] = all.fit_transform(df[i])
```

```python
df
```

```python
df.columns
```

```python
x=df[['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed','Credit_History','Property_Area', 'Income']]
y=df.Application_Status
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y, test_size=0.2, random_state=7)
```

```python
# 7.Create a simple predictive model to assess whether a loan application will be approved or rejected and provide the accuracy score.
# ACCURACY 89% - TEST SET

model = LogisticRegression()
model.fit(x_train,y_train)
predictions = model.predict(x_test)
print(accuracy_score(y_test, predictions)*100)
```

```python
# confusion matrix to identify wrongly accepted applications - 12 applications

pd.crosstab(y_test, predictions)
```
